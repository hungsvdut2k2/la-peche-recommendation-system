**Handle Missing Values Using Random Method**

```python
def fill_in_missing_values(df):
  for column in df:
    if df[column].isna().sum() > 0:
      random_values = np.random.choice(df[df[column].notna()][column], df[column].isna().sum())
      df[column][df[column].isna()] = random_values
  return df
```

```python
fill_in_missing_values(ok_cupid_profiles)
```

```python
import random
import string

def random_char(char_num):
       return ''.join(random.choice(string.ascii_letters) for _ in range(char_num))

ok_cupid_profiles['email'] = [random_char(7)+"@gmail.com" for i in range(ok_cupid_profiles.shape[0])]
```

```python
from datetime import datetime
def get_birthdate(age):
  current_datetime = datetime.now()
  birth_date = f"{current_datetime.year - age}-{current_datetime.month}-{current_datetime.day}"
  birthdate_object = datetime.strptime(birth_date, '%Y-%m-%d').date()
  return birthdate_object
```

```python
ok_cupid_profiles['birth_date'] = [get_birthdate(age) for age in ok_cupid_profiles['age']]
```
